### **Time Series**

```python
#Time Series
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm

prison_df = pd.read_csv('prison.csv')


print(prison_df.head())

trend_data = prison_df.groupby('year')['pris'].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(trend_data['year'], trend_data['pris'], marker='o', linestyle='-', color='b')
plt.title('Trend of Prison Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Prison Population')
plt.grid(True)
plt.show()

correlation_data = prison_df[['pris', 'unem', 'criv', 'crip']]


corr_matrix = correlation_data.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Prison Population and Socio-economic Factors')
plt.show()


sns.pairplot(correlation_data)
plt.show()

prison_ts = trend_data.set_index('year')['pris']


train_data = prison_ts[:int(0.8*len(prison_ts))]
test_data = prison_ts[int(0.8*len(prison_ts)):]


model = ARIMA(train_data, order=(1, 1, 1))  # You can adjust the order for better performance
fitted_model = model.fit()


forecast_steps = len(test_data)
forecast = fitted_model.forecast(steps=forecast_steps)


if len(forecast) != len(test_data):
    forecast = forecast[:len(test_data)]


plt.figure(figsize=(10, 6))
plt.plot(prison_ts, label='Actual', marker='o')
plt.plot(test_data.index, forecast, label='Forecast', linestyle='--', color='r')
plt.title('Prison Population Forecast')
plt.xlabel('Year')
plt.ylabel('Prison Population')
plt.legend()
plt.grid(True)
plt.show()


forecast_df = pd.DataFrame({
     'Year': test_data.index,
     'Actual': test_data.values,
     'Forecast': forecast
})


print(forecast_df)
```

### **K Means**

```python
#K Means
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


df = pd.read_csv('prison.csv')


X = df.iloc[:, [3, 4]].values


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python
# Apply K-Means clustering with 5 clusters
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')

# Plot the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')


plt.title('Clusters of Prison Dataset')
plt.xlabel('Feature 1 (e.g., govelec)')
plt.ylabel('Feature 2 (e.g., black)')
plt.legend()
plt.show()
```

### **Hierarchical Clustering**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import pdist

data = pd.read_csv('prison.csv')
print(data.head(2))


scaler = StandardScaler()
dataNorm = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)


dist_matrix = pdist(dataNorm, metric='euclidean')


hc_complete = linkage(dist_matrix, method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(hc_complete)
plt.title("Dendrogram of Hierarchical Clustering (21BDS0063)")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.grid(True)
plt.show()


k = 3
clusters = fcluster(hc_complete, k, criterion='maxclust')


dataNorm['Cluster'] = clusters


pca = PCA(n_components=2)
pca_result = pca.fit_transform(dataNorm.drop('Cluster', axis=1))
dataNorm['PC1'] = pca_result[:, 0]
dataNorm['PC2'] = pca_result[:, 1]


plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataNorm, x='PC1', y='PC2', hue='Cluster', palette='Set1', s=100)
plt.title("Hierarchical Clustering Scatter Plot (21BDS0063)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

### **MODULE 6**

### **PCA**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.patches as mpatches


dataset = pd.read_csv('prison.csv')
print("First 5 rows of the dataset:")
print(dataset.head())


missing_values = dataset.isnull().sum().sum()
print(f"Total missing values in the dataset: {missing_values}")


numeric_data = dataset.select_dtypes(include=[np.number])


scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)


pca = PCA()
pca_result = pca.fit_transform(scaled_data)


explained_variance = pca.explained_variance_ratio_
print("Explained variance by each component:")
print(explained_variance)


plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='-', color='b')
plt.title('Scree Plot (21BDS0063)')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.xticks(range(1, len(explained_variance) + 1))
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), marker='o', linestyle='-', color='orange')
plt.title('Cumulative Explained Variance (21BDS0063)')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.xticks(range(1, len(explained_variance) + 1))
plt.show()


pca_df = pd.DataFrame(pca_result, columns=[f'PC{i+1}' for i in range(pca_result.shape[1])])
pca_df['ID'] = dataset.index

plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', s=100, color='blue', edgecolor='black')
plt.title('PCA - Individuals (21BDS0063)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
loadings = pca.components_.T
for i, (x, y) in enumerate(zip(loadings[:, 0], loadings[:, 1])):
    plt.arrow(0, 0, x, y, color='r', alpha=0.7, head_width=0.05)
    plt.text(x * 1.1, y * 1.1, dataset.columns[i], color='g', ha='center', va='center')
plt.title('PCA - Variables (21BDS0063)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', s=100, color='blue', edgecolor='black')
for i, (x, y) in enumerate(zip(loadings[:, 0], loadings[:, 1])):
    plt.arrow(0, 0, x, y, color='r', alpha=0.7, head_width=0.05)
    plt.text(x * 1.1, y * 1.1, dataset.columns[i], color='g', ha='center', va='center')
plt.title('PCA - Biplot (21BDS0063)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

### **MODULE 7**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, median_absolute_error, r2_score

file_path = 'prison.csv'
df2 = pd.read_csv(file_path)

X = df2.drop('criv', axis=1)
y = df2['criv']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = median_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)

print("Mean Squared Error:", round(mse, 2))
print("Median Absolute Error:", round(mae, 2))
print("R-squared:", round(r2, 2))
print("Root Mean Squared Error (RMSE):", rmse)
```
